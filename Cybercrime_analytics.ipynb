```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


csdata= pd.read_excel('cybercrime.xlsx')

print(csdata)
```

# New Section

```python
csdf = pd.DataFrame(csdata)
print(csdf)

```

```python
print(csdf.columns)
```

```python
csdf['Total Crimes'] = csdf.iloc[:, 1:].sum(axis=1)
csdf
```

```python
# prompt: Using dataframe csdf: year_wise = csdf.set_index('Year')

csdf_year_wise = csdf.set_index('States')
csdf_year_wise
```

```python
top_5_states = csdf.nlargest(10, 'Total Crimes')
print("]\n Top 5 States with Maximum number of crimes:")
print(top_5_states)
```

```python
csdf.describe()
```

```python
cmatrix =csdf.iloc[:, 1:].corr()
cmatrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(csdf.set_index('States').iloc[:, 1:].corr(), annot=True, cmap='rainbow')
plt.title('CHeatmap of values over the years')
plt.xlabel('States')
plt.ylabel('year')
plt.show()
```

```python
correlation_matrix = csdf.iloc[:, 1:].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add title and labels
plt.title('Correlation Heatmap')
plt.xlabel('States')
plt.ylabel('Year')
plt.show()
```
