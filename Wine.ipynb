```python
from os import WIFSIGNALED
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn  import datasets
from scipy.stats import f_oneway

wine = datasets.load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target
df.head()
```

```python
# prompt: generate the histogram using wine data

import matplotlib.pyplot as plt
df.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python

columns = ['alcohol','malic_acid','ash','alcalinity_of_ash','magnesium','total_phenols','flavanoids','nonflavanoid_phenols','proanthocyanins','color_intensity','hue','od280/od315_of_diluted_wines','proline'] # Adjusted column names to match DataFrame

# Create individual plots
for column in columns:
    plt.figure(figsize=(6,4))  # Create a new figure for each plot
    sns.displot(df[column], kde=True, bins=20)  # Plot the distribution of the column
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    plt.show()

```

```python
column_for_ANOVA=['alcohol','malic_acid','ash']
anova_results = f_oneway(*[df[column] for column in column_for_ANOVA])
print(anova_results)
```

**ANOVA INTERPRETETATION**



*   The ANOVA result indicate a statistical significat different among the means of the variables(Alcohol,Malic.acid,ash) tested
*   Statistic: The F-statstic is very large(10183.02),suggesting singnification variablility between groups relative to within groups.


*  P-Value: The p-value is 0.0, Which is less than any conventional singnificance level(e.g 0.05).This indicates strong evidence against the null hypothesis suggesting that at h=least of the variables(Alcohol,Malic.acid.Ash) significantly influences the outcomes.

*  In Summary the ANOVA test shows that there are signification difference in at least one of the variable (Alcohol, Malic.acid,Ash) across differnet group ofcategories in your dataset
